```python
# Activation Function
def relu(x):
    return np.maximum(0, x)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)
```

```python
# Forward Pass
def dense(a_in, W, b, g):
    z = np.dot(a_in, W) + b
    return g(z)
```

```python
def forward_pass(x, W1, b1, W2, b2, W3, b3):
    a1 = dense(x, W1, b1, relu)
    a2 = dense(a1, W2, b2, relu)
    a3 = dense(a2, W3, b3, softmax)
    return a3
```

```python
# Fitness Function (Cross-Entropy Loss with Accuracy)
def fitness_function(chromosome, X_train, y_train, input_neurons, H1_neurons, H2_neurons, output_neurons):
    # Decode chromosome
    W1_len = input_neurons * H1_neurons
    b1_len = H1_neurons
    W2_len = H1_neurons * H2_neurons
    b2_len = H2_neurons
    W3_len = H2_neurons * output_neurons
    b3_len = output_neurons

    # Extract weights and biases from chromosome
    W1 = chromosome[:W1_len].reshape(input_neurons, H1_neurons)
    b1 = chromosome[W1_len:W1_len + b1_len]
    W2 = chromosome[W1_len + b1_len:W1_len + b1_len + W2_len].reshape(H1_neurons, H2_neurons)
    b2 = chromosome[W1_len + b1_len + W2_len:W1_len + b1_len + W2_len + b2_len]
    W3 = chromosome[W1_len + b1_len + W2_len + b2_len:W1_len + b1_len + W2_len + b2_len + W3_len].reshape(H2_neurons, output_neurons)
    b3 = chromosome[-b3_len:]

    # Forward pass
    y_pred = forward_pass(X_train, W1, b1, W2, b2, W3, b3)
    y_pred_labels = np.argmax(y_pred, axis=1)

    # Compute fitness as accuracy
    accuracy = accuracy_score(y_train, y_pred_labels)
    return accuracy
```

```python
# Generate Random Population
def initialize_population(pop_size, chromosome_length):
    return np.random.uniform(-0.5, 0.5, size=(pop_size, chromosome_length))
```

```python
# Selection
def select_population(population, fitness_scores, num_selected):
    sorted_indices = np.argsort(fitness_scores)[::-1]
    return population[sorted_indices[:num_selected]]
```

```python
# Crossover
def crossover(parent1, parent2, crossover_rate=0.5):
    mask = np.random.rand(len(parent1)) < crossover_rate
    return np.where(mask, parent1, parent2)
```

```python
# Mutation
def mutate(chromosome, mutation_rate=0.01):
    mutation_mask = np.random.rand(len(chromosome)) < mutation_rate
    chromosome[mutation_mask] += np.random.uniform(-0.1, 0.1, size=np.sum(mutation_mask))
    return chromosome
```

```python
# Generate Next Generation
def next_generation(selected_population, num_offspring, crossover_rate=0.5, mutation_rate=0.01):
    next_gen = []
    for _ in range(num_offspring):
        parent1, parent2 = np.random.choice(len(selected_population), 2, replace=False)
        offspring = crossover(selected_population[parent1], selected_population[parent2], crossover_rate)
        offspring = mutate(offspring, mutation_rate)
        next_gen.append(offspring)
    return np.array(next_gen)
```

```python
# Genetic Algorithm
def genetic_algorithm(X_train, y_train, input_neurons, H1_neurons, H2_neurons, output_neurons, pop_size, generations, num_selected):
    # Define chromosome length
    chromosome_length = (input_neurons * H1_neurons + H1_neurons +
                         H1_neurons * H2_neurons + H2_neurons +
                         H2_neurons * output_neurons + output_neurons)

    # Initialize population
    population = initialize_population(pop_size, chromosome_length)

    for generation in range(generations):
        # Evaluate fitness
        fitness_scores = np.array([fitness_function(ind, X_train, y_train,
                                                     input_neurons, H1_neurons, H2_neurons, output_neurons)
                                    for ind in population])

        # Select top individuals
        selected_population = select_population(population, fitness_scores, num_selected)

        # Generate new offspring
        new_generation = next_generation(selected_population, pop_size - num_selected)

        # Update population
        population = np.vstack((selected_population, new_generation))

        # Print best fitness
        best_fitness = np.max(fitness_scores)
        print(f"Generation {generation + 1}: Best Fitness = {best_fitness:.4f}")

    # Return the best chromosome
    final_fitness_scores = np.array([fitness_function(ind, X_train, y_train,
                                                       input_neurons, H1_neurons, H2_neurons, output_neurons)
                                      for ind in population])
    best_index = np.argmax(final_fitness_scores)
    return population[best_index]
```

```python
# Define activation functions
def relu(z):
    return np.maximum(0, z)
```

```python
def softmax(z):
    exps = np.exp(z - np.max(z, axis=1, keepdims=True))  # Stability adjustment
    return exps / np.sum(exps, axis=1, keepdims=True)
```

```python
def dense(a_in, W, b, g):
    z = np.dot(a_in, W) + b  # Matrix multiplication for batch input
    a_out = g(z)  # Apply activation function element-wise
    return a_out
```

```python
# Sequential network function
def sequential(X):
    a1 = dense(X, W1, b1, relu)   # Hidden layer 1 with ReLU activation
    a2 = dense(a1, W2, b2, relu)  # Hidden layer 2 with ReLU activation
    a3 = dense(a2, W3, b3, softmax)  # Output layer with Softmax activation
    return a3
```

```python
# Loss function (categorical cross-entropy)
def categorical_crossentropy(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred + 1e-9)) / 4
```

```python
# Accuracy calculation
def accuracy(y_true, y_pred):
    pred_classes = np.argmax(y_pred, axis=1)
    true_classes = np.argmax(y_true, axis=1)
    return np.mean(pred_classes == true_classes)
```

##Notes: One-Hot Encoding:

Since your model is multi-class classification, we convert the target labels into one-hot encoded vectors. For example, for 4 classes, if a sample's label is 2, it will be converted to [0, 0, 1, 0].
This is required for the softmax output layer to produce probabilities for each class.


```python
def encode_chromosome(W1, b1, W2, b2, W3, b3):
    # Flatten and concatenate all weights and biases into a single 1D array
    chromosome = np.concatenate([
        W1.flatten(), b1.flatten(),
        W2.flatten(), b2.flatten(),
        W3.flatten(), b3.flatten()
    ])
    return chromosome

# Example usage
chromosome = encode_chromosome(W1, b1, W2, b2, W3, b3)
```

```python
def generate_population(pop_size, chromosome_length):
    # Generate 'pop_size' random chromosomes of length 'chromosome_length'
    population = np.random.uniform(
        low=-0.5, high=0.5, size=(pop_size, chromosome_length)
    )
    return population

# Example usage
population_size = 20
chromosome_length = len(chromosome)  # Total parameters of the MLP
population = generate_population(population_size, chromosome_length)
```

```python
def decode_chromosome(chromosome, input_neurons, H1_neurons, H2_neurons, output_neurons):
    # Unroll the chromosome into weights and biases
    W1_len = input_neurons * H1_neurons
    b1_len = H1_neurons
    W2_len = H1_neurons * H2_neurons
    b2_len = H2_neurons
    W3_len = H2_neurons * output_neurons
    b3_len = output_neurons

    W1 = chromosome[:W1_len].reshape((input_neurons, H1_neurons))
    b1 = chromosome[W1_len:W1_len + b1_len]
    W2 = chromosome[W1_len + b1_len:W1_len + b1_len + W2_len].reshape((H1_neurons, H2_neurons))
    b2 = chromosome[W1_len + b1_len + W2_len:W1_len + b1_len + W2_len + b2_len]
    W3 = chromosome[W1_len + b1_len + W2_len + b2_len:W1_len + b1_len + W2_len + b2_len + W3_len].reshape((H2_neurons, output_neurons))
    b3 = chromosome[W1_len + b1_len + W2_len + b2_len + W3_len:]

    return W1, b1, W2, b2, W3, b3
```

```python
# Fitness function (e.g., accuracy or negative loss)
def fitness_function(chromosome, X_train, y_train, input_neurons, H1_neurons, H2_neurons, output_neurons):
    W1, b1, W2, b2, W3, b3 = decode_chromosome(chromosome, input_neurons, H1_neurons, H2_neurons, output_neurons)
    predictions = []
    for x in X_train:
        # Forward pass through the network
        a1 = np.dot(x, W1) + b1
        a2 = np.dot(a1, W2) + b2
        output = np.dot(a2, W3) + b3
        predictions.append(np.argmax(output))  # Get the predicted class
    accuracy = np.mean(predictions == y_train)  # Compare predictions to ground truth
    return accuracy
```

```python
def select_population(population, fitness_scores, num_selected):
    # Sort by fitness score (higher is better) and select the top individuals
    sorted_indices = np.argsort(fitness_scores)[::-1]
    selected_population = population[sorted_indices[:num_selected]]
    return selected_population

# Example usage
num_selected = 10
fitness_scores = np.array([fitness_function(ind, X_train, y_train, 6, 8, 6, 4) for ind in population])
selected_population = select_population(population, fitness_scores, num_selected)
```
